```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
iris_data = pd.read_csv('/kaggle/input/iris/Iris.csv')
```

```python
X = iris_data.drop("Species", axis=1)
Y = iris_data["Species"]
```

```python
Y_numeric = pd.Categorical(Y).codes
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_numeric, test_size=0.2, random_state=40)
```

```python
# Logistic Regression
logistic_model = LogisticRegression(random_state=120)
logistic_model.fit(X_train, Y_train)
logistic_predictions = logistic_model.predict(X_test)
logistic_accuracy = accuracy_score(Y_test, logistic_predictions)
```

```python
# K-Nearest Neighbors
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, Y_train)
knn_predictions = knn_model.predict(X_test)
knn_accuracy = accuracy_score(Y_test, knn_predictions)
```

```python
# Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, Y_train)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(Y_test, rf_predictions)
```

```python
# Display results
print("\n========= Model Evaluation ==========")
print(f"Logistic Regression Accuracy: {logistic_accuracy:.4f}")
print(f"K-Nearest Neighbors Accuracy: {knn_accuracy:.4f}")
print(f"Random Forest Classifier Accuracy: {rf_accuracy:.4f}")
```

```python
# Compare classification reports
print("\n========= Logistic Regression Classification Report ==========")
print(classification_report(Y_test, logistic_predictions))

print("\n========= K-Nearest Neighbors Classification Report ==========")
print(classification_report(Y_test, knn_predictions))

print("\n========= Random Forest Classifier Classification Report ==========")
print(classification_report(Y_test, rf_predictions))
print("===========================================")
```
